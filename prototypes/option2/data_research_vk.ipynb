```python
univercity_df = pd.DataFrame(df.groupby(['university_id','university_name'])['record_id'].count()).sort_values('record_id',ascending=False).reset_index()
univercity_df
```

```python
univercity_df.to_csv('univercity.csv', sep = ',', index=False)
```

```python
faculty_df = pd.DataFrame(df.groupby(['faculty_name'])['record_id'].count()).sort_values('record_id',ascending=False).reset_index()
faculty_df
```

```python
faculty_df.to_csv('faculty.csv', sep = ',', index=False)
```

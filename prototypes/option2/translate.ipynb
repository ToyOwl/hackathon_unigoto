```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('faculty.csv', delimiter = ',', low_memory=False)
```

```python
df
```

```python
df['faculty_name_fixed'] = df['faculty_name'].str.replace(r'\(.*\)', ' ',regex=True).str.replace(r'«.»', ' ',regex=True).str.replace(r'("|«|»)', ' ',regex=True).str.replace(r'\s{2,}', ' ',regex=True).str.strip()
```

```python
df.nunique()
```

```python
df
```

```python
df.to_csv('faculty_fixed.csv', sep = ',', index=False)
```

```python
df1= pd.DataFrame(df['faculty_name_fixed'].unique()).rename(columns={0:'faculty_name_fixed'})
```

```python
df1
```

```python
df1.iloc[0,0]
```

```python
prev_transl_df = pd.read_csv('faculty_fixed_translated.csv', delimiter = ',', low_memory=False)
```

```python
prev_transl_df
```

```python
prev_transl_df.set_index('faculty_name_fixed', inplace=True)
```

```python
prev_transl_df
```

```python
joined_df = df1.join(prev_transl_df, on='faculty_name_fixed', how='left')
#df1 = df1.join(df_comments_summary, on='Ticket key', how='left')
```

```python
joined_df
```

```python
joined_df.info()
```

```python
joined_df['faculty_name_ru'].isna().sum()
```

```python
lines2translate = list(joined_df[joined_df['faculty_name_ru'].isna()].index)
```

```python

# 259m52s for 41187 rows (2.64 per second)
#df1 = pd.DataFrame(columns = ['faculty_name','cnt', 'faculty_name_fixed','faculty_name_ru', 'is_different'])
#df2 = pd.DataFrame(columns = ['faculty_name_fixed','faculty_name_ru', 'is_different'])

for cnt, i in enumerate(lines2translate):
#    f_name = df.loc[i,'faculty_name']
#    cnt = df.loc[i,'record_id']
    f_name_fixed = joined_df.loc[i,'faculty_name_fixed']
    f_name_ru = GoogleTranslator(source='auto', target='ru').translate(f_name_fixed)
    is_different = int(f_name_fixed != f_name_ru)
    
    joined_df.loc[i,'faculty_name_ru'] = f_name_ru
    joined_df.loc[i,'is_different'] = is_different
#    df2.loc[len(df2)] = [f_name_fixed, f_name_ru, is_different]
    if cnt % 1000 ==0:
        print(cnt)
    
```
